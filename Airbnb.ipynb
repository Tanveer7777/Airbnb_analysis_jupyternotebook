# handling missing values

```python
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'review per month' : 0, 'last review':df['last review'].min()},inplace=True)
```

df.fillna(...)
Calls the method that fills missing (NaN) values in the DataFrame.

{'review per month': 0, 'last review': df['last review'].min()}
A dictionary specifying the fill value for each column:

Fill 'review per month' NaNs with 0.

Fill 'last review' NaNs with the minimum date from that column (i.e., the earliest review date).

inplace=True
Means the changes are applied directly to the DataFrame df, rather than returning a modified copy.

```python
df.dropna(subset=['NAME','host name'],inplace=True)
```

fillna() keeps all rows by replacing missing values.

dropna() removes rows that have missing values.

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=["license","house_rules"], errors="ignore")
```

```python
df.head(5)
```

```python
#remove $ sign from price and service key for avoiding errors and convert into float
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$]','',regex=True).astype(float)
```

```python
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# visualization
# What is the distribution of listing price ?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='g')
plt.title("Distribution Of Listing Price",fontsize=25)
plt.xlabel("price $",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

The histogram shows a fairly even distribution of listing prices
across differnt price ranges,indicating no particular concentration
of listing in any specific price range.

The KDE line helps visualize this even spread more cleraly,
confirming that the dataset contains listing with a wide variety of price

# What is the relationship between price and room type ?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price",hue="room type",data=df)
plt.title("Price VS Room Type",fontsize=25)
plt.xlabel("Room Type",fontsize=15)
plt.ylabel("Price ($)",fontsize=15)
plt.legend(title='Room Type')
plt.show()
```

# how has the number of reviews change over time ?

```python
df['last reviews']=pd.to_datetime(df['last review'])
reviews_over_time=df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title("Number of reviews over time",fontsize=25)
plt.xlabel("Data")
plt.ylabel("Number of Reviews")
plt.show()
           
```
